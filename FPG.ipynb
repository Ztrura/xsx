```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

import pandas as pd

df = pd.read_csv('./xsx2.csv',encoding='utf-8-sig')
```

```python
df
```

```python
# for i in range(df.shape[0]):
#     if df.loc[i,'满意程度'] <= 4:
#         df.loc[i,'满意程度'] = "低"
#     elif df.loc[i,'满意程度'] > 4 and df.loc[i,'满意程度'] <=6:
#         df.loc[i,'满意程度'] = "中"
#     else:
#         df.loc[i,'满意程度'] = "高"
for i in range(df.shape[0]):
    if df.loc[i,'满意程度'] <= 5:
        df.loc[i,'满意程度'] = "低"
    else:
        df.loc[i,'满意程度'] = "高"
```

```python
df["满意程度"] = df["满意程度"].apply(lambda x:str(x))
```

```python
df
```

```python
# apply one-hot encoding
ohe_df = pd.get_dummies(df.apply(pd.Series.explode)).sum(level=0)
```

```python
ohe_df
```

```python
frequent_itemsets = fpgrowth(ohe_df, min_support=0.6, use_colnames=True)

# print frequent itemsets
frequent_itemsets = frequent_itemsets[frequent_itemsets["itemsets"].apply(lambda x: len(x)>1)]
frequent_itemsets = frequent_itemsets.reset_index(drop=True)
frequent_itemsets
```

```python
indexs = set()
for i in range(len(frequent_itemsets)):
#     print(result.loc[i]["attribute"])
    for j in range(len(frequent_itemsets.loc[i]["itemsets"])):
        if "满意程度" in list(frequent_itemsets.loc[i]["itemsets"])[j]:
#             print(i)
            indexs.add(i)
frequent_itemsets.loc[indexs]
```

```python
# apply fp-growth algorithm
frequent_itemsets = fpgrowth(ohe_df, min_support=0.55, use_colnames=True)

# print frequent itemsets
frequent_itemsets = frequent_itemsets[frequent_itemsets["itemsets"].apply(lambda x: len(x)>0)]
frequent_itemsets = frequent_itemsets.reset_index(drop=True)
# frequent_itemsets
```

```python
asrules = association_rules(frequent_itemsets, metric='support', min_threshold=0.6, support_only=False)
asrules
```
