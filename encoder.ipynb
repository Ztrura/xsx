```python
df = pd.read_csv('./xsx.csv', encoding='utf-8-sig')
```

```python
df
```

```python
for i in range(df.shape[0]):
    if df.loc[i,'人均年收入']=='(空)':
        df.loc[i,'人均年收入'] = '空'
    elif int(df.loc[i,'人均年收入']) < 10000:
        df.loc[i,'人均年收入'] = '一万以下'
    elif int(df.loc[i,'人均年收入']) >= 10000 and int(df.loc[i,'人均年收入']) < 100000:
        df.loc[i,'人均年收入'] = '一万~十万'
    elif int(df.loc[i,'人均年收入']) >= 100000 and int(df.loc[i,'人均年收入']) < 200000:
        df.loc[i,'人均年收入'] = '十万~二十万'
    elif int(df.loc[i,'人均年收入']) >= 200000:
        df.loc[i,'人均年收入'] = '二十万及以上'
```

```python
for name in header:
    class_mapping = {label: idx for idx, label in enumerate(set(df[name]))}
    with open('./%s_dict.json' % name, 'w' ,encoding='utf-8-sig') as f:
        json.dump(class_mapping, f, indent = 2, ensure_ascii=False)
    df[name] = df[name].map(class_mapping)
```
